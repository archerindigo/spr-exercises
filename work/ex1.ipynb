# Exercise 1

Suppose we have three colored boxes r (red), b (blue), and g (green).
Box r contains 3 apples, 4 oranges, and 3 limes, box b contains 1 apple, 1
orange and no limes, and box g contains 3 apples, 3 oranges, and 4 limes.
If a box is chosen at random with probabilities p(r)=0.2, p(b)=0.2, p(g)=0.6,
and a piece of fruit is removed from the box with equal probability of
selecting any of the items in the box, then what is the probability of
selecting an apple? If we observe that the selected fruit is in fact an
orange, what is the probability that it came from the green box?

將所有拎到蘋果的機率相加
p(apple) = p(apple, r) + p(apple, b) + p(apple, g)

(use product rule)

$
\begin{align}
p(apple) &= p(apple|r) p(r) + p(apple|b) p(b) + p(apple|g) p(g) \\
         &= 3/10 * 0.2 + 1/2 * 0.2 + 3/10 * 0.6 \\         
         &= 0.34
\end{align}
$
         
If we observe that the selected fruit is in fact an
orange, what is the probability that it came from the green box?


用Bayes rule

$
\begin{align}
p(g|orange) &= \frac{p(orange|g) p(g)}{p(orange)} \\
\\
p(orange) &= 0.4 * 0.2 + 0.5 * 0.2 + 0.4 * 0.6 \\
          &= 0.42 \\
\\
p(g|orange) &= 0.3 * 0.6 / 0.42 \\
            &= 0.429
\end{align}
$

# Part 1: Matrix Inversion
Define a matrix and a vector and compute the result of a matrix-vector multiplication.
Invert the matrix.

```python
a = np.arange(1,10)
a.shape = 3,3
a[2,1] = 2
b = np.arange(3)  # [0, 1, 2]
multiply = a @ b   # Multiply @
#multiply = np.matmul(a, b)  # old way
print(multiply) 

print(np.linalg.inv(a)) # Use inv() to inverse a matrix
```
